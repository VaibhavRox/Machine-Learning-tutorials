```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
```

```python
#Fit is used to train the model
model.fit(X_train,Y_train)
```

```python
#To find the accuracy of model, pass X_test, model creates predicted values
#Compares the predicted values to actual values Y_test
model.score(X_test,Y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
model.predict([digits.data[67]])
```

```python
model.predict(digits.data[0:5])
```

#### Check where the model fails, using confusion matrix

```python
y_predicted=model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_predicted)
cm
```

```python
#visualize the matrix using seaborn
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
